```python
import numpy as np
import scipy as sp
import scipy.stats as st
import sympy as sm
import matplotlib.pyplot as plt
from scipy.special import gamma
```

```python
x = np.linspace(-5, 5, 100)
y = gamma(x)
```

```python
plt.plot(x, y, 'b', alpha=0.8, label='gamma(x)')
plt.xlim(-5, 5)
plt.ylim(-25, 25)
plt.grid()
plt.xlabel('x')
plt.legend(loc='lower right')
plt.show()
```

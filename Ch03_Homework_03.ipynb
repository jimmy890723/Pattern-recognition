```python
import numpy as np
import scipy as sp
import scipy.stats as st
import sympy as sm
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import yfinance as yf
sb.set_theme()
```

```python
#
# Top10 Companies
#
top10= {
    'MSFT':    'Microsoft',
    'AAPL':    'Apple',
    'GOOGL':    'Alphabet (Google)',
    #'2222.SR': 'Saudi Aramco',
    'AMZN':    'Amazon',
    'TSLA':    'Tesla',
    'FB':      'Meta (Facebook)',
    'NVDA':    'NVIDIA',
    'BRK-A':   'Berkshire Hathaway',
    'TSM':     'TSMC',
    'TCEHY':   'Tencent',
    }
```

```python
top10L= list(top10.keys())
top10L
```

```python

tks= yf.Tickers(['MSFT', 'AAPL', 'GOOGL', #'2222.SR',
                 'AMZN', 'TSLA', 'FB', 'NVDA', 'BRK-A','TSM','TCEHY'])
tks

tkPrice= tks.history(period='5Y')

#  X= X(t)
tkPrice.Close.plot(ylabel='$USD')
tkPrice.Close.plot(logy= True, ylabel='$USD')

# Histogram
tkPrice.Close.hist()
tkPrice.Close.hist(orientation= 'horizontal')
```

```python
top10Stocks= [
    (2.482E+03,  336.63),
    (2.712E+03,  160.24),
    (1.887E+03, 2910.61), 
    #(6.966E+03,  34.65), 
    (1.779E+03,  3561.57), 
    (1.15E+03,  1136.99), 
    (902.57	,   338.03), 
    (814.286,   333.76), 
    (617.989,   425260), 
    (607.549,   119.28), 
    (562.014,   58.67),  
    ]

#top10Stocks= np.array(top10Stocks)[:,-1]*1e-3 # unit= Trillion
#top10Stocks

top10Stocks= np.array(top10Stocks)[:,0]/ np.array(top10Stocks)[:,1]/1e3 # unit= Trillion
top10Stocks
```

```python
Top10Companies= tkPrice.Close.loc[:,top10L]*top10Stocks
Top10Companies.plot(#logy=True, 
       title= 'Top10 largest Companies, Market Cap',
       ylabel= 'Trillion ($10^{12}$) Dallars')
```
